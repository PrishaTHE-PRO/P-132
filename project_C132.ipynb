```python
import csv
import pandas as pd
rows = []
df=pd.read_csv('star_with_gravity.csv')
df.head()
```

```python
mass=df['Mass'].tolist()
radius=df['Radius'].tolist()
gravity=df['Gravity'].tolist()
mass.sort()
gravity.sort()
radius.sort()
import matplotlib.pyplot as plt
plt.plot(radius,mass)
plt.title('radius and mass of the star')
plt.xlabel('radius')
plt.ylabel('mass')
plt.show()
```

```python
plt.plot(gravity,mass)
plt.title('mass and gravity of the star')
plt.xlabel('mass')
plt.ylabel('gravity')
plt.show()
```

```python
plt.scatter(radius,mass)
plt.title('radius and mass of the star')
plt.xlabel('radius')
plt.ylabel('mass')
plt.show()
```
